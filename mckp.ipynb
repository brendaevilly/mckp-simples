```python
def mckp(grupos, capacidade):
    # dp[peso] = melhor valor possível com esse peso
    dp = [-1] * (capacidade + 1)
    dp[0] = 0  # começar com peso 0 e valor 0

    # percorre cada grupo
    for grupo in grupos:
        novo_dp = [-1] * (capacidade + 1)

        for peso_atual in range(capacidade + 1):
            if dp[peso_atual] != -1:  # só pesa válido
                # para cada item do grupo, testar escolher
                for peso_item, valor_item in grupo:
                    novo_peso = peso_atual + peso_item

                    if novo_peso <= capacidade:
                        novo_valor = dp[peso_atual] + valor_item

                        # atualiza melhor valor daquele novo peso
                        if novo_valor > novo_dp[novo_peso]:
                            novo_dp[novo_peso] = novo_valor

        dp = novo_dp  # avança para o próximo grupo

    return max(dp)  # melhor valor alcançável


if __name__ == "__main__":
    # (peso, valor)
    grupos_exemplo = [
        [(2, 6), (3, 10), (4, 12)],
        [(1, 4), (3, 7), (4, 9)], 
        [(2, 5), (3, 8)]       
    ]
    capacidade_exemplo = 7

    valor= mckp(grupos_exemplo, capacidade_exemplo)
    print("Melhor valor:", valor)
```

```python
def mckp(grupos, capacidade):
    # Inicializa o vetor dp com tamanho igual à capacidade + 1
    dp = []

    for peso in range(capacidade + 1):
        if peso == 0:
            # Com peso 0, o melhor valor possível é 0 e não selecionamos nenhum item
            dp.append((0, []))
        else:
            # Para qualquer outro peso, ainda não há solução válida
            dp.append((-1, None))


    # Percorre cada grupo
    for g, grupo in enumerate(grupos):
        novo_dp = [(-1, None)] * (capacidade + 1)

        for peso_atual in range(capacidade + 1):
            valor_atual, escolhas_atual = dp[peso_atual]

            if valor_atual >= 0:  # estado válido
                # Tenta pegar cada item do grupo
                for peso_item, valor_item in grupo:
                    novo_peso = peso_atual + peso_item

                    if novo_peso <= capacidade:
                        novo_valor = valor_atual + valor_item

                        # Atualiza se for melhor
                        if novo_valor > novo_dp[novo_peso][0]:
                            novo_dp[novo_peso] = (
                                novo_valor,
                                escolhas_atual + [(g, peso_item, valor_item)]
                            )

        dp = novo_dp  # avança para o próximo grupo

    # Melhor solução ao final
    melhor_peso = max(range(capacidade + 1), key=lambda p: dp[p][0])
    melhor_valor, itens_escolhidos = dp[melhor_peso]

    return melhor_valor, itens_escolhidos

if __name__ == "__main__":
    # (peso, valor)
    grupos_exemplo = [
        [(2, 6), (3, 10), (4, 12)],
        [(1, 4), (3, 7), (4, 9)],
        [(2, 5), (3, 8)]
    ]
    capacidade_exemplo = 7

    print("\n=== GRUPOS DE ITENS ===")
    for i, grupo in enumerate(grupos_exemplo):
        print(f"Grupo {i}:")
        for p, v in grupo:
            print(f"  - peso {p}, valor {v}")
        print()

    melhor_valor, itens = mckp(grupos_exemplo, capacidade_exemplo)

    print("\n=== RESULTADO FINAL ===")
    print(f"Melhor valor encontrado: {melhor_valor}\n")

    print("Itens escolhidos:")
    for g, p, v in itens:
        print(f"  - Grupo {g}: peso {p}, valor {v}")
```
